## Conversiones

### 1.Conversión decimal -> binario











*   Paso 1: Hacer la división entera del número decimal entre 2
*   Paso 2: Guardar el modúlo de de la división entre 2
*   Paso 3: Repetir los pasos 1 y 2 hasta obtener 0 en el cociente
*   Paso 4: Invertir los módulos guardados
*   Paso 5: Concatenar los módulos para obtener el número final
*   Paso 6: Encapsular el algoritmo en una función

```python
def dectobin(numero_decimal):
    numero_decimal=int(numero_decimal)
    modulos = []

    if numero_decimal == 0:
        modulos.insert(0,'0')
    else:
        while numero_decimal != 0:
            modulos.insert(0,str(numero_decimal%2))
            numero_decimal //=2
    numero_binario = int("".join(modulos))
    return numero_binario
```

### 2.Conversión binario -> decimal



*   Paso 1: Determinar el formato del número binario
*   Paso 2: Recorrer digitos
*   Paso 3: Saber en que posición estoy
*   Paso 4: Multiplicar cada digito por 2^posicion
*   Paso 5: Sumar valores de las multiplicaciones
*   Paso 6: Encapsular el algoritmo en una función

```python
def bintodec(numero_binario):
    numero_binario = str(numero_binario)
    posicion = len(numero_binario)-1

    numero_decimal=0
    for posicion,digito_binario in enumerate(numero_binario[::-1]):
        numero_decimal += int(digito_binario)*2**posicion 

    return numero_decimal
```

### 3.Conversión decimal -> octal



*   Paso 1: Hacer la división entera del número decimal entre 8
*   Paso 2: Guardar el modúlo de de la división entre 8
*   Paso 3: Repetir los pasos 1 y 2 hasta obtener un número menor a 8 en el cociente
*   Paso 4: Invertir los módulos guardados
*   Paso 5: Concatenar los módulos para obtener el número final
*   Paso 6: Encapsular el algoritmo en una función

```python
def dectooct(numero_decimal):
    numero_decimal=int(numero_decimal)
    modulos = []

    if numero_decimal == 0:
        modulos.insert(0,'0')
    else:
        while numero_decimal != 0:
            modulos.insert(0,str(numero_decimal%8))
            numero_decimal //=8
    numero_octal = int("".join(modulos))
    return numero_octal
```

### 4.Conversión octal -> decimal



*   Paso 1: Determinar el formato del número octal
*   Paso 2: Recorrer digitos
*   Paso 3: Saber en que posición estoy
*   Paso 4: Multiplicar cada digito por 8^posicion
*   Paso 5: Sumar valores de las multiplicaciones
*   Paso 6: Encapsular el algoritmo en una función

```python
def octtodec(numero_octal):
    numero_octal = str(numero_octal)
    posicion = len(numero_octal)-1

    numero_decimal=0
    for posicion,digito_octal in enumerate(numero_octal[::-1]):
        numero_decimal += int(digito_octal)*8**posicion 

    return numero_decimal
```

### 5.Conversión decimal -> hexadecimal



*   Paso 1: Hacer la división entera del número decimal entre 16
*   Paso 2: Guardar el modúlo de de la división entre 16
*   Paso 3: Repetir los pasos 1 y 2 hasta obtener un número menor a 16 en el cociente
*   Paso 4: Invertir los módulos guardados
*   Paso 5: Convertir los módulos que correspondan a letras
*   Paso 6: Concatenar los módulos para obtener el número final
*   Paso 7: Encapsular el algoritmo en una función

```python
def dectohex(numero_decimal):
    numero_decimal = int(numero_decimal)
    modulos = []
    equivalencias = {"10": "A", "11": "B", "12": "C", "13": "D", "14": "E", "15": "F"}
    
    if numero_decimal == 0:
        modulos.insert(0,'0')
    else:
        while numero_decimal != 0:
            modulos.insert(0,str(numero_decimal%16))
            numero_decimal //=16

    for i in range(len(modulos)):
        if modulos[i] in equivalencias:
            modulos[i] = equivalencias[modulos[i]]
    
    numero_hexadecimal = str("".join(modulos))
    return numero_hexadecimal
```

### 6.Conversión hexadecimal -> decimal



*   Paso 1: Determinar el formato del número octal
*   Paso 2: Recorrer digitos
*   Paso 3: Convertir las letras a sus valores equivalentes
*   Paso 4: Saber en que posición estoy
*   Paso 5: Multiplicar cada digito por 8^posicion
*   Paso 6: Sumar valores de las multiplicaciones
*   Paso 7: Encapsular el algoritmo en una función

```python
def hextodec(numero_hexadecimal):
    numero_hexadecimal = str(numero_hexadecimal).upper()
    posicion = len(numero_hexadecimal)-1

    equivalencias = {"A": "10", "B": "11", "C": "12", "D": "13", "E": "14", "F": "15"}

    numero_decimal=0
    for posicion,digito_octal in enumerate(numero_hexadecimal[::-1]):
        if digito_octal in equivalencias:
            digito_octal = equivalencias[digito_octal]
        numero_decimal += int(digito_octal)*16**posicion 

    return numero_decimal
```

### 7.Conversión octal -> binario




*   Paso 1: Aplicar la función octal a decimal
*   Paso 2: Aplicar la función decimal a binario
*   Paso 3: Encapsular en una función

```python
def octtobin(numero_octal):
    numero_binario=dectobin(octtodec(numero_octal))
    return numero_binario
```

### 10.Conversión binario -> hexadecimal

## Operaciones

### 1.Suma

```python
def sumbinarios(num1,num2):
    suma = bintodec(num1)+bintodec(num2)
    return dectobin(suma)
```

```python
def sumoctales(num1,num2):
    suma = octtodec(num1)+octtodec(num2)
    return dectooct(suma)
```

```python
def sumhexadecimal(num1,num2):
    suma = hextodec(num1)+hextodec(num2)
    return dectohex(suma)
```

### 2.Resta

```python
def resbinarios(num1,num2):
    resta = bintodec(num1)-bintodec(num2)
    return dectobin(resta)
```

```python
def resoctales(num1,num2):
    resta = octtodec(num1)-octtodec(num2)
    return dectooct(resta)
```

```python
def reshexadecimal(num1,num2):
    resta = hextodec(num1)-hextodec(num2)
    return dectohex(resta)
```

### 3.Multipicación

```python
def mulbinarios(num1,num2):
    multiplicacion = bintodec(num1)*bintodec(num2)
    return dectobin(multiplicacion)
```

```python
def muloctales(num1,num2):
    multiplicacion = octtodec(num1)*octtodec(num2)
    return dectooct(multiplicacion)
```

```python
def mulhexadecimal(num1,num2):
    multiplicacion = hextodec(num1)*hextodec(num2)
    return dectohex(multiplicacion)
```

### 4.División

```python
def divbinarios(num1,num2):
    division = bintodec(num1)//bintodec(num2)
    return dectobin(division)
```

```python
def divoctales(num1,num2):
    division = octtodec(num1)//octtodec(num2)
    return dectooct(division)
```

```python
def divhexadecimal(num1,num2):
    division = hextodec(num1)//hextodec(num2)
    return dectohex(division)
```

## Complemento a 2


```python
def complementoados(num,bits):
    if num<0:
        complemento = dectobin(num*-1)
    else:
        num = dectobin(num)
        complemento = (str('0'*(bits-len(str(num))))+str(num)).replace('0','2').replace('1','0').replace('2','1')
        complemento = int("".join(complemento))
        complemento = sumbinarios(complemento,1)
    return complemento
```
